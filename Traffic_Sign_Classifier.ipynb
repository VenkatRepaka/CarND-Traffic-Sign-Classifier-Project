## Generate Fake Data

### Function to add gaussian noise to image

````python
def add_noise(image_to_add_noise):
    sigma = np.random.randint(10, 20)
    return image_to_add_noise + np.random.normal(0., sigma, image_to_add_noise.shape).astype(np.uint8)
````

### Function to change orientation of image (tilt the image)

````python
def change_orientation(image_to_change):
    rotation_angle = np.random.randint(10, 20)
    (h, w) = image_to_change.shape[:2]
    center = (w/2, h/2)
    M = cv2.getRotationMatrix2D(center, rotation_angle, 1.0)
    return cv2.warpAffine(image_to_change, M, (w, h))
````

### Function to shift the centre of the image 

````python
def shift_center(image_to_shift):
    x_Pos, y_pos = np.random.randint(-4, 4, 2)
    M = np.float32([[1,0,x_Pos],[0,1,y_pos]])
    (h, w) = image_to_shift.shape[:2]
    return cv2.warpAffine(image_to_shift, M, (w, h))
````

### Function to change perspective of image

````python
def chnage_perspective(image_to_perspective):
    rows,cols = image_to_perspective.shape
    low_x, low_y = np.random.randint(2, 8, 2)
    high_x, high_y = np.random.randint(26, 31, 2)
    pts1 = np.float32([[low_x,low_y],[high_y,low_x],[low_x,high_y],[high_x,high_y]])
    pts2 = np.float32([[0,0],[32,0],[0,32],[32,32]])

    M = cv2.getPerspectiveTransform(pts1,pts2)

    return cv2.warpPerspective(image_to_perspective,M,(rows,cols))
````

### Function to shrink image

````python
def shrink_image(image_to_shrink):
    shrink_expand = np.random.randint(1, 3)
    print(shrink_expand)
    if shrink_expand == 1:
        change_size = np.random.randint(27, 31)
    else:
        change_size = np.random.randint(34, 37)
    ratio = change_size/image_to_shrink.shape[1]
    new_dim = (change_size, int(image_to_shrink.shape[0]*ratio))
    return cv2.resize(image_to_shrink, new_dim, interpolation = cv2.INTER_AREA)
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
````

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

def LeNetWithDropout(x, keep_prob):
    mu = 0
    sigma = 0.1
    
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean=mu, stddev=sigma))
    conv1_B = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID')
    
    conv1 = tf.nn.relu(conv1)
    conv1 = tf.nn.dropout(conv1, keep_prob=keep_prob)
    
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean=mu, stddev=sigma))
    conv2_B = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID')
    
    conv2 = tf.nn.relu(conv2)
    conv2 = tf.nn.dropout(conv2, keep_prob=keep_prob)
    
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    flattened_after_pool = flatten(conv2)
    
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 120), mean=mu, stddev=sigma))
    fc1_B = tf.Variable(tf.zeros(120))
    fc1 = tf.add(tf.matmul(flattened_after_pool, fc1_W), fc1_B)
    
    fc1 = tf.nn.relu(fc1)
    fc1 = tf.nn.dropout(fc1, keep_prob=keep_prob)
    
    fc2_W = tf.Variable(tf.truncated_normal(shape=(120, 84), mean=mu, stddev=sigma))
    fc2_B = tf.Variable(tf.zeros(84))
    fc2 = tf.add(tf.matmul(fc1, fc2_W), fc2_B)
    
    fc2 = tf.nn.relu(fc2)
    fc2 = tf.nn.dropout(fc2, keep_prob=keep_prob)
    
    fc3_W = tf.Variable(tf.truncated_normal(shape=(84, n_classes), mean=mu, stddev=sigma))
    fc3_B = tf.Variable(tf.zeros(n_classes))
    logits = tf.add(tf.matmul(fc2, fc3_W), fc3_B)
    
    return logits
    
````

````python
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, n_classes)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
````

````python
learning_rate = 0.001
logits = LeNetWithDropout(x, keep_prob)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss = tf.reduce_mean(cross_entropy)
training = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_predictin = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

epochs = 40
BATCH_SIZE = 156

def evaluate(X_data, y_data):
    n_example = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, n_example, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_predictin, feed_dict={x:batch_x, y: batch_y, keep_prob:1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy/n_example
````

````python
from sklearn.utils import shuffle

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    n_example = len(X_train_gray)
    
    print('Started training')
    for i in range(epochs):
        # X_train_gray, y_train = shuffle(X_train_gray, y_train)
        X_train_gray, y_train = shuffle(X_total_train, y_total_train)
        X_train_gray_expanded = [img[:, :, np.newaxis] for img in X_train_gray]
        for offset in range(0, n_example, BATCH_SIZE):
            batch_x, batch_y = X_train_gray_expanded[offset:offset+BATCH_SIZE], y_train[offset:offset+BATCH_SIZE]
            sess.run(training, feed_dict={x: batch_x, y: batch_y, keep_prob:0.7})
        X_valid_gray_expanded = [img[:, :, np.newaxis] for img in X_valid_gray]
        training_accuracy = evaluate(X_train_gray_expanded, y_train)
        validation_accuracy = evaluate(X_valid_gray_expanded, y_valid)
        print("Epoch {} .. ".format(i+1))
        print("Training accuracy = {:.3f}".format(training_accuracy))
        print("Validation accuracy = {:.3f}".format(validation_accuracy))
        
    saver.save(sess, "./lenet")
    print("Model saved.")
````
